```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
housing_data = pd.read_csv('Housing.csv')
```

```python
# Displaying the first few rows
print(housing_data.head())
```

```python
# Check for missing values
missing_values = housing_data.isnull().sum()
print('Missing values in each column:', missing_values)
```

```python
# Summary statistics for numerical columns
describe_data = housing_data.describe()
print('Summary statistics for numerical columns:', describe_data)
```

```python
# Convert 'yes'/'no' columns to boolean
bool_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in bool_columns:
    housing_data[col] = housing_data[col].map({'yes': True, 'no': False})
```

```python
# Convert 'furnishingstatus' to categorical type
housing_data['furnishingstatus'] = housing_data['furnishingstatus'].astype('category')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='bedrooms', data=housing_data, palette='viridis')
plt.title('Distribution of Bedrooms')
plt.show()
```

```python
#Plot the distribution of the 'area' variable using a histogram
plt.figure(figsize=(10, 6))
plt.hist(housing_data['area'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Area in the Housing Dataset')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```
